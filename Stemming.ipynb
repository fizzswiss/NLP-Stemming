```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
for word in words:
    print(word+'----->'+stemming.stem(word))
```

```python
stemming.stem('congratulations')
```

```python
stemming.stem('sitting')
```

```python
from nltk.stem import RegexpStemmer

reg_stem=RegexpStemmer('ing$|s$|e$|able$',min=4)
```

```python
reg_stem.stem('eating')
```

```python
reg_stem.stem('ingeating')
```

```python
reg_stem.stem('storms')
```

```python
reg_stem.stem('eatable')
```

```python
reg_stem.stem('morninge')
```

```python
from nltk.stem import SnowballStemmer
```

```python
snball_stem=SnowballStemmer(language='english')
```

```python
snball_stem.stem('mornings')
```

```python
for sn_word in words:
    print(sn_word+'--->'+snball_stem.stem(sn_word))
```

```python
snball_stem.stem('fairly')
```

```python
snball_stem.stem('rapidfire')
```

```python
snball_stem.stem('passerby')
```
